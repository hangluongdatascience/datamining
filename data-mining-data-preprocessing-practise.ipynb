**1. Delete:** 
df.dropna(axis=0, inplace=True) # Remove rows with missing values

**2. Mean/Median/Mode Imputation: **

* Replace missing values with the mean, median, or mode of the observed data for that variable.
    df['column_name'].fillna(df['column_name'].mean(), inplace=True)
* Regression Imputation: Predict missing values based on the relationship with other variables through regression analysis.
* K-nearest neighbors (KNN) Imputation: Estimate missing values based on the values of neighboring data points.

**3. Flagging Missing Values:**

df['column_name_missing'] = df['column_name'].isnull().astype(int)

**4. Domain-specific Methods:**

df['column_name'].fillna('Other', inplace=True)

**5. Model-Based Imputation**

from sklearn.tree import DecisionTreeRegressor # Train a model to predict missing values
X = df.dropna().drop('column_name', axis=1)
y = df.dropna()['column_name']

dt_model = DecisionTreeRegressor()
dt_model.fit(X, y)

missing_values = df[df['column_name'].isnull()].drop('column_name', axis=1)
predicted_values = dt_model.predict(missing_values)

df.loc[df['column_name'].isnull(), 'column_name'] = predicted_values

# **3. Noise Handling**

```python
# Group by 'payment_method' and count unique transaction_id and customer_id
result = df.groupby('Payment Method').agg({
    'Transaction ID': 'nunique',
    'Customer ID': 'nunique',
    'Transaction Amount': 'sum' ,
     'Quantity':'sum'
}).reset_index()

# Rename columns for clarity
result.columns = ['payment_method', 'unique_transaction_count', 'unique_customer_count','sum_transaction_amount','quantity']
result
```

Paypal, bank transfer, credit card, debit card are have the same percentage of contribution (this rates makes me confused due to its beauty ?!) 

```python
# Group by 'Product Category' and count unique transaction_id and customer_id
category_info = df.groupby(['Payment Method','Product Category']).agg({
    'Transaction ID': 'nunique',
    'Customer ID': 'nunique',
    'Transaction Amount': 'sum' ,
     'Quantity':'sum'
}).reset_index()

# Rename columns for clarity
category_info.columns = ['payment_method','category', 'unique_transaction_count', 'unique_customer_count','sum_transaction_amount','quantity']
category_info
```
